# Salary calculation using polynomial regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
# import dataset into df data frame
# previous database ds_salaries.csv, not perfect to this regression
df = pd.read_csv('salary.csv')
```

```python
# prepare graph with seaborn
sns.set_theme()
sns.set_style("whitegrid")
graph = sns.scatterplot(x="YearsExperience", y="Salary", data = df)
graph.set(ylabel='Salary USD', xlabel='Experience lvl')
plt.show()
```

```python
# linear regression for values in pyplot
reg = linear_model.LinearRegression()
reg.fit(df[['YearsExperience']], df['Salary'])

plt.xlabel('Experience')
plt.ylabel('Salary USD')

plt.scatter(df['YearsExperience'], df['Salary'])

xekseni = df['YearsExperience']
yekseni = reg.predict(df[['YearsExperience']])
plt.plot(xekseni, yekseni, color='green', label = 'linear regression')
plt.legend()
plt.show()
reg.coef_
reg.intercept_
```

```python
# Adapt our x value to fit the polynomial function above
# So => 1, x , x^2 (N=2)
```

```python
# Call PolynomialFeatures function to create a polynomial regression object
# Specify the degree (N) of the polynomial when calling this function
polynomial_regression = PolynomialFeatures(degree = 4)

x_polynomial = polynomial_regression.fit_transform(df[['YearsExperience']])
```

```python
# Here fit the x_polynomial and y axes by creating reg object, which is regression model object,
# and calling its fit method
# so we train our regression model with data:
reg = linear_model.LinearRegression()
reg.fit(x_polynomial, df['Salary'])
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(df['YearsExperience'], y_head, color ='red', label = 'polynomial regression')
plt.legend()

# Scatter data as point and see if it fits polynomial regression:
plt.scatter(df['YearsExperience'], df['Salary'])

plt.show()
```
